```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv("city_data.csv")
ride_data = pd.read_csv("ride_data.csv")

# Combine the data into a single dataset

combined_data = city_data.merge(ride_data, on = 'city')
combined_data

# Display the data table for preview
```

```python
color = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
```

```python
average_fare = combined_data.groupby("city")["fare"].mean()
```

```python
total_rides = combined_data.groupby("city")["ride_id"].count()
```

```python
total_drivers = combined_data.groupby("city")["driver_count"].mean()
```

```python
city_type = city_data.set_index("city")['type']

city_type.value_counts()
```

```python
pyber_df = pd.DataFrame({"Average Fare": average_fare, "Total Number of Rides Per City": total_rides, 
                        "Total Drivers Per City": total_drivers, "Type of City": city_type})

pyber_df
```

```python
urban = pyber_df.loc[pyber_df["Type of City"]== "Urban"]
suburban = pyber_df.loc[pyber_df["Type of City"]== "Suburban"]
rural = pyber_df.loc[pyber_df["Type of City"]== "Rural"]
urban
```

```python
plt.scatter(urban["Total Number of Rides Per City"], 
            urban["Average Fare"], 
            color = "gold", edgecolors="black",
            alpha = 0.4,
            s = urban["Total Drivers Per City"]*6, 
            label = "Urban", linewidth = 1.5)

plt.scatter(suburban["Total Number of Rides Per City"], 
            suburban["Average Fare"], 
            color = "lightcoral", edgecolors="black",
            alpha = 0.4,
            s = suburban["Total Drivers Per City"]*6, 
            label = "Suburban", linewidth = 1.5)

plt.scatter(rural["Total Number of Rides Per City"], 
            rural["Average Fare"], 
            color = "lightblue", edgecolors="black", 
            alpha = 0.4,
            s = rural["Total Drivers Per City"]*6, 
            label = "Rural", linewidth = 1.5)


plt.title("Pyber Ride Sharing Data")
plt.xlabel("Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(42, 45,"Note: Circle size correlates with driver count per city.")


plt.legend(loc= "upper right")


plt.show()
```

```python
urban_type = combined_data.loc[combined_data["type"]== "Urban"]
urban_fare = urban_type["fare"].sum()

suburban_type = combined_data.loc[combined_data["type"]== "Suburban"]
suburban_fare = suburban_type["fare"].sum()

rural_type = combined_data.loc[combined_data["type"]== "Rural"]
rural_fare = rural_type["fare"].sum()


total_fare =[urban_fare,suburban_fare,rural_fare]
labels =["Urban","Suburban","Rural"]
colors = ["gold","lightcoral","lightblue"]
plt.pie(total_fare,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True)
plt.title("Percentage of Fares By City Type")
sns.set
plt.show()
```

```python
Total_urban_rides = pyber_df.loc[pyber_df["Type of City"] == "Urban"]
urban_rides = Total_urban_rides["Total Number of Rides Per City"].sum()

Total_suburban_rides = pyber_df.loc[pyber_df["Type of City"] == "Suburban"]
suburban_rides = Total_suburban_rides["Total Number of Rides Per City"].sum()

Total_rural_rides = pyber_df.loc[pyber_df["Type of City"] == "Rural"]
rural_rides = Total_rural_rides["Total Number of Rides Per City"].sum()

total_rides = [urban_rides,suburban_rides,rural_rides]
labels =["Urban","Suburban","Rural"]
colors = ["gold","lightcoral","lightblue"]
plt.pie(total_rides,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True)
plt.title("Percentage of Rides by City Type")
sns.set
plt.show()
```

```python
Total_urban_drivers = pyber_df.loc[pyber_df["Type of City"] == "Urban"]
urban_drivers = Total_urban_drivers["Total Drivers Per City"].sum()

Total_suburban_drivers = pyber_df.loc[pyber_df["Type of City"] == "Suburban"]
suburban_drivers = Total_suburban_drivers["Total Drivers Per City"].sum()

Total_rural_drivers = pyber_df.loc[pyber_df["Type of City"] == "Rural"]
rural_drivers = Total_rural_drivers["Total Drivers Per City"].sum()

total_rides = [urban_drivers,suburban_drivers,rural_drivers]
labels =["Urban","Suburban","Rural"]
colors = ["gold","lightcoral","lightblue"]
plt.pie(total_rides,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True)
plt.title("Percentage of Drivers by City Type")
sns.set
plt.show()
```
